```python
import pandas as pd
import json
```

```python
import numpy as np
```

```python
from pandas import json_normalize
```

```python
file_path = "E:\hospital-response\chatbot\intents.json"
```

```python
with open(file_path, 'r') as json_file:
    json_data = json.load(json_file)
```

```python
column_names = ['message', 'intent']
```

```python
df = pd.DataFrame(columns=column_names)
```

```python
df
```

```python
json_data
```

```python
response = {}
```

```python
for intent in json_data['intents']:
    for pattern in intent['patterns']:
        new_row = {
            'message' : pattern,
            'intent' : intent['tag']
        }
        df = df.append(new_row, ignore_index = True)
            
```

```python
df
```

```python
for intent in json_data['intents']:
    for respo in intent['responses']:
        new_key = intent['tag']
        new_value = respo

        if new_key in response:
            response[new_key].append(new_value)
        else:
            response[new_key] = [new_value]
```

```python
for key in response.keys():
    print(key)
```

```python
new_df = df.sample(frac=1).reset_index(drop=True)
```

```python
new_df
```

```python
new_df.to_csv('data.csv')
```

```python
new_df.value_counts('intent')
```

```python
new_df['label'] = new_df.intent.map({
'symptoms':               0,
'fallback' :              1,
'service_completion':     2,
'greeting' :              3,
'addmission_enqiry' :     4,
'session_reschedule':     5,
'session_cancelation':    6,
'book_session':           7,
'payment':                8,
'feedback':               9,
'medical_services':       10,
'contact_location':       11
})
```

```python
import tensorflow as tf
from tensorflow import keras
import tensorflow_hub as hub
import tensorflow_text as text
```

```python
new_df
```

```python
new_df.to_csv('data.csv')
```

```python
encoder = "C:/Users/nitis/Desktop/main_model"
preprocess = "C:/Users/nitis/Desktop/model"
```

```python
test_text = ["Hi, what are you doing", "where are you going rita", "I have become death, destroyer of worlds"]
```

```python
preprocess_model = hub.KerasLayer(preprocess)
```

```python
model_inputs = preprocess_model(test_text)
```

```python
model_inputs
```

```python
bert_model = hub.KerasLayer(encoder)
```

```python
embeddings = bert_model(model_inputs)
```

```python
embeddings.keys()
```

```python
embeddings['pooled_output'][0].numpy()
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(new_df['message'], new_df['label'], test_size=0.2, stratify=new_df['label'])
```

```python
X_train
```

```python
X_test
```

```python
text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')
preprocessing_layer = hub.KerasLayer(preprocess, name='preprocessing')
encoder_inputs = preprocessing_layer(text_input)
encoder = hub.KerasLayer(encoder, trainable=True, name='BERT_encoder')
outputs = encoder(encoder_inputs)
net = outputs['pooled_output']
net = tf.keras.layers.Dropout(0.1)(net)
net = tf.keras.layers.Dense(12, activation='softmax', name='classifier')(net)
model = tf.keras.Model(inputs = [text_input], outputs = [net])

model.summary()
```

```python
model.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-5),
              loss = 'SparseCategoricalCrossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=1)
```

```python
model.evaluate(X_test, y_test)
```

```python
intent = ['symptoms', 'fallback', 'service_completion', 'greeting', 'addmission_enqiry', 'session_reschedule', 'session_cancelation', 'book_session', 'payment', 'feedback']
```

```python
def predict_intent(text):
    x = model.predict([text])
    ind = np.argmax(x)
    return intent[ind]
```
