```python
import pandas as pd
import json
```

```python
import numpy as np
```

```python
file_path = "/content/symptoms.json"
```

```python
with open(file_path, 'r') as json_file:
    json_data = json.load(json_file)
```

```python
column_names = ['text', 'symp']
```

```python
df = pd.DataFrame(columns=column_names)
```

```python
df
```

```python
json_data
```

```python
for categ in json_data.keys():
    for text in json_data[categ]:
        new_row = {
            'text' : text,
            'symp' : categ
        }
        df = df.append(new_row, ignore_index = True)
```

```python
df
```

```python
new_df = df.sample(frac=1).reset_index(drop=True)
```

```python
new_df
```

```python
new_df.to_csv('data.csv')
```

```python
new_df.value_counts('symp')
```

```python
new_df['label'] = new_df.symp.map({
  'symptom' :       0,
  'None':           1
})
```

```python
new_df
```

```python
import tensorflow as tf
from tensorflow import keras
import tensorflow_hub as hub
import tensorflow_text as text
```

```python
new_df.to_csv('data.csv')
```

```python
encoder = "https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4"
preprocess = "https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3"
```

```python
test_text = ["Hi, what are you doing", "where are you going rita", "I have become death, destroyer of worlds"]
```

```python
preprocess_model = hub.KerasLayer(preprocess)
```

```python
model_inputs = preprocess_model(test_text)
```

```python
model_inputs
```

```python
bert_model = hub.KerasLayer(encoder)
```

```python
embeddings = bert_model(model_inputs)
```

```python
embeddings.keys()
```

```python
embeddings['pooled_output'][0].numpy()
```

```python
from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(new_df['text'], new_df['label'], test_size=0.2, stratify=new_df['label'])
```

```python
X_train.shape
```

```python
X_test
```

```python
text_input = tf.keras.layers.Input(shape=(), dtype=tf.string, name='text')
preprocessing_layer = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3", name='preprocessing')
encoder_inputs = preprocessing_layer(text_input)
encoder = hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4", trainable=True, name='BERT_encoder')
outputs = encoder(encoder_inputs)
net = outputs['pooled_output']
net = tf.keras.layers.Dropout(0.1)(net)
net = tf.keras.layers.Dense(1, activation='sigmoid', name='classifier')(net)
model = tf.keras.Model(inputs = [text_input], outputs = [net])

model.summary()
```

```python
loss = 'binary_crossentropy'
metrics = ['accuracy']
```

```python
init_lr = 1e-5
optimizer = tf.keras.optimizers.AdamW(learning_rate=init_lr)
```

```python
model.compile(optimizer=optimizer,
                         loss=loss,
                         metrics=metrics)
```

```python
num_epochs = 2
batch_size = 32
total_samples = 3120
steps_per_epoch = total_samples // batch_size
model.fit(X_train, y_train, epochs=num_epochs, steps_per_epoch=steps_per_epoch)
```

```python
model.evaluate(X_test, y_test)
```

```python
def predict_intent(text):
    x = model.predict([text])
    return "symptom" if x <0.5 else "none"
```

```python
predict_intent("please provide me with status of my appointment")
```

```python
y_predic = model.predict(X_test)
```

```python
y_pred = [0 if x < 0.5 else 1 for x in y_predic]
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
